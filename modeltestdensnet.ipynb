```python
import pandas as pd
import os

# Define directories
train_dir = '/content/drive/MyDrive/trainrice'
val_dir = '/content/drive/MyDrive/valrice'
test_dir = '/content/drive/MyDrive/testrice'

# Function to create a dataframe for images
def create_df(img_dir):
    img_paths = [os.path.join(img_dir, img) for img in os.listdir(img_dir) if img.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tiff'))]
    df = pd.DataFrame({'filename': img_paths, 'class': 0})  # Use 0 for all class labels
    return df


train_df = create_df(train_dir)
val_df = create_df(val_dir)
test_df = create_df(test_dir)

print(f'Train DataFrame: {train_df.shape}')
print(f'Validation DataFrame: {val_df.shape}')
print(f'Test DataFrame: {test_df.shape}')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Data augmentation for training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    zoom_range=0.2,
    horizontal_flip=True
)

# Only rescaling for validation and testing
val_test_datagen = ImageDataGenerator(rescale=1./255)

# Use class_mode=None since there's only one class (rice)
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    x_col='filename',            # Image filenames
    y_col=None,                  # No labels for the images
    target_size=(256, 256),      # Target size for ResNet50
    batch_size=32,
    class_mode=None              # No class labels since we only have one class
)

val_generator = val_test_datagen.flow_from_dataframe(
    dataframe=val_df,
    x_col='filename',
    y_col=None,                  # No labels for the images
    target_size=(256, 256),
    batch_size=32,
    class_mode=None
)

test_generator = val_test_datagen.flow_from_dataframe(
    dataframe=test_df,
    x_col='filename',
    y_col=None,                  # No labels for the images
    target_size=(256, 256),
    batch_size=32,
    class_mode=None,
    shuffle=False                # Don't shuffle test data
)
```

```python
conv_l = DenseNet121(weights='imagenet', include_top=False, input_shape=(256, 256, 3))
conv_l.trainable = False

# Build the model
model = Sequential([
    conv_l,
    BatchNormalization(),
    Flatten(),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dense(1, activation='sigmoid')  # One output class (1: rice leaf, 0: not rice leaf)
])

# Compile the model
model.compile(
    optimizer=Adam(learning_rate=0.01),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Early stopping to prevent overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

def custom_generator(generator):
    while True:  # Loop indefinitely
        for x in generator:  # Unpack images only
            yield x, np.ones((x.shape[0], 1))  # Create labels of 1

```

```python
# Train the model with custom generator
history = model.fit(
    custom_generator(train_generator),
    validation_data=custom_generator(val_generator),
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    validation_steps=val_generator.samples // val_generator.batch_size,
    epochs=25,
    callbacks=[early_stopping]
)
```

```python
# Create a custom generator for the test set that yields images with labels of 1 (rice leaves)
def custom_test_generator(generator):
    while True:
        for x in generator:  # Unpack images only
            yield x, np.ones((x.shape[0], 1))  # Create labels of 1 for all test images

# Evaluate the model using the custom test generator
test_loss, test_accuracy = model.evaluate(
    custom_test_generator(test_generator),
    steps=test_generator.samples // test_generator.batch_size
)

# Print the test accuracy
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
